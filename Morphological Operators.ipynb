# <span style = "color:coral;font-size:40px"> Morphological Operators </span>

Morphological transformations are some simple operations based on the image shape. It is normally performed on binary images. It needs two inputs, one is our original image, second one is called structuring element or kernel which decides the nature of operation. Two basic morphological operators are Erosion and Dilation. Then its variant forms like Opening, Closing, Gradient etc also comes into play

***

### Import the required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

### Erosion

Erodes away boundaries of foregroung objects. Works best when foregroung is light color (preferrably white) and background is dark

```python
img = load_img()
kernel = np.ones((5,5))
erosion1 = cv2.erode(img, kernel, iterations = 1)
```

```python
display_img(erosion1)
```

```python
img = load_img()
kernel = np.ones((5,5))
erosion5 = cv2.erode(img,kernel,iterations = 4)
```

```python
display_img(erosion5)
```

### Opening

Opening is erosion followed by dilation. Useful in removing background noise!

```python
img = load_img()
white_noise = np.random.randint(low = 0, high = 2, size = (600,600))
white_noise
```

```python
white_noise = white_noise* 255
```

```python
white_noise.shape
```

```python
img.shape
```

```python
noise_img = white_noise+ img
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

### Closing

Usefull in removing noise from foreground objects, such as black dots on top of the white text.

```python
img = load_img()
```

```python
black_noise = np.random.randint(low = 0, high = 2, size = (600,600))
black_noise
```

```python
black_noise = black_noise * -255
```

```python
black_noise_img = img + black_noise
```

```python
black_noise_img[black_noise_img == -255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```

### Morphological Gradient

Difference between dilation and erosion of an image

```python
img = load_img()
display_img(img)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```
