```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
packet = cv2.imread('packet.jpg')
```

```python
plt.imshow(packet)
```

```python
shelf = cv2.imread('shelf.jpg')
```

```python
display(shelf)
```

### Brute Force Detection with ORB Descriptors

```python
# Initiate ORB detector
orb = cv2.ORB_create()

# find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(packet,None)
kp2, des2 = orb.detectAndCompute(shelf,None)

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors.
matches = bf.match(des1,des2)

# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 25 matches.
packet_matches = cv2.drawMatches(packet,kp1,shelf,kp2,matches[:25],None,flags=2)
```

```python
display(packet_matches)
```
