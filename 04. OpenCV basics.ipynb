### Import necessary Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Read 'image1.jpg' with opencv

```python
img = cv2.imread('image1.jpg')
```

```python
img
```

### Plot the image with matplotlib

```python
plt.imshow(img)
```

### Convert the image to RGB

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

### Check the shape of the image

```python
img_rgb.shape
```

### Resize the image to (400,300)

```python
resized_img = cv2.resize(img_rgb,(400,300))
```

```python
plt.imshow(resized_img)
```

### Resize the image by ratio. Use 0.5 height and width

```python
w_ratio=1
h_ratio=0.5
```

```python
ratioed_img = cv2.resize(img_rgb,(0,0),img_rgb,h_ratio,w_ratio)
```

```python
plt.imshow(ratioed_img)
```

### Flip the image along central x axis

```python
x_img = cv2.flip(img_rgb,0)
```

```python
plt.imshow(x_img)
```

### Flip the image along central y axis 

```python
y_img = cv2.flip(img_rgb,1)
```

```python
plt.imshow(y_img)
```
