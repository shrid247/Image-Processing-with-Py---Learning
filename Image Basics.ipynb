### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Open 'image2.jpg' from the data folder using open cv and display it in the notebook. Make sure to correct for the RGB order

```python
img = cv2.imread('image2.jpg')
```

##### Before RGB correction

```python
plt.imshow(img)
```

### After correction

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Draw an empty RED rectangle around toms face and display the image in the notebook

```python
cv2.rectangle(img_rgb,pt1=(300,300),pt2=(750,750),color=(255,0,0),thickness=5)
plt.imshow(img_rgb)
```

### Draw a text named 'TOM' above toms face

```python
font=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText(img_rgb,text='TOM',org=(570,800),fontFace=font,fontScale=2, color=(0,0,250), thickness=3,lineType=cv2.LINE_AA)
plt.imshow(img_rgb)
```

### Do the same steps for 'Jerry'

```python
cv2.rectangle(img_rgb,pt1=(100,500),pt2=(300,700),color=(255,0,0),thickness=5)
plt.imshow(img_rgb)
```

```python
font=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText(img_rgb,text='Jerry',org=(100,750),fontFace=font,fontScale=2, color=(0,0,0), thickness=3,lineType=cv2.LINE_AA)
plt.imshow(img_rgb)
```

### Convert the image to grayscale

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

### Display the image in a larger size

```python
plt.figure(figsize=(10,10))
plt.imshow(img_gray,cmap='gray')
```
